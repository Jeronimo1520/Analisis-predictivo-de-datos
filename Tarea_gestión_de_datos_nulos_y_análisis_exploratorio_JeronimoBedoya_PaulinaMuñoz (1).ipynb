# Imputación de horsepower con la media

```python
df['horsepower'].fillna(value=df['horsepower'].mean(), inplace=True)
df.info()
```

# Estadisticas descriptivas

```python
df['origin'] = df['origin'].astype('object')
df["cylinders"] = df["cylinders"].astype("object")
df["model_year"] = df["model_year"].astype("object")
df.info()
```

# Diagramas

## Diagramas de barras para variables categoricas

```python
cat_vars = df.select_dtypes(include=['object']).columns
for var in cat_vars:
    df[var].value_counts().plot(kind='bar')
    plt.title(var)
    plt.show()
```

Se puede analizar que la mayoría de registros tiene un origin 1, y tienen 4 cilindros, también existe una pequeña mayoría que tienen cilindros impares.

Se puede ver que no existe una gran diferencia entre los valores de la variable model_year, pero la mayoría de registros pertenecen al año 73.

## Histograma para variables numericas

- Origen vs cilíndros: En este caso vemos lo que pasa cuando se hace este tipo de gráfico con dos variables categóricas, esto no es recomendable ni ideal ya que se puede producir gráficos sin sentido o no muy claros de analizar.

- Origin vs displacement: Se puede decir que los autos de origen 1 tienen mayor desplazamiento por tanqueada, con una gran diferencia en valores respecto a los autos de los origines 2 y 3.

- Origin vs horsepower: No existe un gran diferencia entre magnitudes de los valores de las medianas de los autos de los 3 origenes, siendo mayores los de origen 1 y los de origen 2 y 3 son muy parecidos, cabe también recalcar que el origen en el único que presenta outliers.

- Origin vs weight: Se puede ver que la mediana de los autos con origen 1 es mayor a los autos de los distintos origenes, siendo los de origen 2 y 3 más parecidos, lo que nos puede decir que los autos más pesados pertenecen a el origen 1. No se encuentrán valores atípicos en ningún caso.

- Origin vs Acceleration: La aceleración entre los autos de los 3 orígenes casi no cambia, son muy parecidas, respecto a su mediana, pero se puede recalcar que el rango de valores de aceleración de la región 1 es más amplio respecto a las otras.
